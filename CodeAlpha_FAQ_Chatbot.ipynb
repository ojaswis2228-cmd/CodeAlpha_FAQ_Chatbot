```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_question"] = df["question"].apply(clean_text)

df[["question", "clean_question"]]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(
    stop_words="english",
    ngram_range=(1, 2)
)

faq_vectors = vectorizer.fit_transform(df["clean_question"])

print("Model updated successfully")
```

```python
def find_best_answer(user_question):
    cleaned_question = clean_text(user_question)

    user_vector = vectorizer.transform([cleaned_question])

    similarity_scores = cosine_similarity(
        user_vector,
        faq_vectors
    )[0]

    best_index = similarity_scores.argmax()
    best_score = similarity_scores[best_index]

    if best_score < 0.20:
        return {
            "answer": "I could not find a confident answer. Please rephrase your question.",
            "matched_question": None,
            "similarity_score": best_score
        }

    return {
        "answer": df.iloc[best_index]["answer"],
        "matched_question": df.iloc[best_index]["question"],
        "similarity_score": best_score
    }
```

```python
result = find_best_answer("How much attendance do I need?")

print("Matched Question:", result["matched_question"])
print("Answer:", result["answer"])
print("Similarity Score:", round(result["similarity_score"], 2))
```
